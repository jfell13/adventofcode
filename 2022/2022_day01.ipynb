```python
total_carry = {}
starting_input = open('inputs/2022_day01.txt','r')
elf_num = 1
carry_sum = 0
for line in starting_input:
    if line == '\n':
        total_carry[elf_num] = carry_sum
        carry_sum = 0
        elf_num += 1
    else:
        line = int(line.strip('\n'))
        carry_sum = carry_sum + line
total_carry
```

```python
max_energy = max(total_carry.values())
max_energy
```

```python
list(total_carry.values()).index(max_energy)
```

```python
energies = list(total_carry.values())
energies.sort(reverse=True)
```

```python
energies = energies[0:3]
new_total = 0
for i in energies:
    new_total += i
new_total
```
